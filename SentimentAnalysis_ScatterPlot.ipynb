```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tweets.csv")
# data.set_index("Net Score", inplace=True, drop=False, append=False)
data
```

```python
def mark(val):
    if val == 0:
        return "o"
    elif val > 0:
        return "+"
    else:
        return "*"

def color(val):
    if val == 0:
        return "tab:blue"
    elif val > 0:
        return "tab:green"
    else:
        return "tab:red"
    
data["marker"] = data["Net Score"].apply(mark)
data["color"] = data["Net Score"].apply(color)
data
```
